```python
class Login:

    def __init__(self):
        self.userid = ''
        self.password = ''
        self.menu()


    def menu(self):
        user_input = input("""

        1. press 1 for new ragstraton
        2. press 2 for login 
        """)

        if user_input == '1':
            self.new_ragistration()
        elif user_input == '2':
            self.new_login()
        else:
            exit()
            

    def new_ragistration(self):
        user_id = input("Enter your id:")
        self.userid = user_id
        user_pd = input("Enter your password:")
        self.password = user_pd

        print("Sucessfully ragister")
        self.menu()

    def new_login(self):
        user_id = input("Enter your id:")
        user_pd = input("Enter your password:")

        if user_id == self.userid and user_pd == self.password:
            print(f"Login successfullly, Welcome {self.userid}")
        else:
            print("wrong login parameters")

        self.menu()
        
```
